```python
import pandas as pd
import numpy as np
```

```python
i_cols = ['movie id', 'movie title' ,'release date','video release date', 'IMDb URL', 'unknown', 'Action', 'Adventure',
    'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
    'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
items = pd.read_csv(fileloc+'/ml-100k/u.item', sep='|', names=i_cols,encoding='latin-1')
```

```python
movie_content=items[['movie id','movie title','Action', 'Adventure',
    'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
    'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']]
movie_content
```

```python
content_vector = pd.read_csv(fileloccsv)
content_vector
```

```python
def top5movies(content_vectors,movie_content):
    val_list = []
    for i in content_vectors:
        val = content_vectors[i].max()
        val_list.append(val)

    val_df=pd.DataFrame(val_list)
    val_df.columns = ['max_id']
    result = (val_df.sort_values(by='max_id', ascending=False)).head(5)
    movie_titles = movie_content[movie_content['movie id'].isin (result.index)][['movie title','Action', 'Adventure',
    'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy','Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']]

    genre_cols = movie_titles.columns.drop('movie title')
    movie_titles['Active_Genres'] = movie_titles[genre_cols].apply(lambda row: row.index[row == 1].tolist(), axis=1)
    result = movie_titles[['movie title', 'Active_Genres']]
    return result
```

```python
movie_titles = top5movies(content_vector,movie_content)
movie_titles
```
