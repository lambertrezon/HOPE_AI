```python
#pandas is a library and it contains the functions related to the database
import pandas as pd
```

```python
dataset=pd.read_csv("Salary_Data.csv")#dataset is a variable where we are using pandas function to read the CSV file and load it
```

```python
dataset #Printing the data present behind dataset variable
```

```python
independent=dataset[["YearsExperience"]] #Loading input data. Assigning specific column to variable.
independent
```

```python
dependent=dataset[["Salary"]] #Loading output data.Assigning specific column to variable.
dependent
```

```python
from sklearn.model_selection import train_test_split # loading train_test_split function from sklearn.model_selection
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30,random_state=0)
#calling train_test_split parameterized function with inputs as input variable, output variable & load test set with 30%
```

```python
from sklearn.linear_model import LinearRegression # import LinearRegression machine learning function from sklearn.linear_model library
regressor=LinearRegression() # loading LinearRegression function to variable regressor
regressor.fit(X_train,y_train) # substituting input and output variable for the machine learning function
```

```python
weight=regressor.coef_ # y=wx+b, this line will calculate the weight of data using regressor.coef_ function
weight # display variable value
```

```python
bais=regressor.intercept_ # y=wx+b, this line will calculate the bais of data using regressor.intercept_ function
bais# display variable value
```

```python
y_pred=regressor.predict(X_test) # Training is completed now we are passing test value
```

```python
from sklearn.metrics import r2_score # Evaluation matrix: sum of r square is being imported here as function
r_score=r2_score(y_test,y_pred) # passing the test value and model predicted value
```

```python
r_score # output the r2_score value
```

```python
import pickle # This function is specifically meant for saving the trained model
filename="finalized_model_linear.sav" # assigning the filename
```

```python
pickle.dump(regressor,open(filename,'wb')) # writing the trained model in binary to the filename with extension
```

```python
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb')) # reading the saved model in binary and loading it to variable
result=loaded_model.predict([[13]]) # passing input as 13 and see the result ( passing 13 yrs as input and getting the salaruy prediction for 13 yrs)
```
