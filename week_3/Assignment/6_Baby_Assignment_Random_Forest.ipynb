```python
import pandas as pd
dataset = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/HOPE AI/WEEK-3-ML/1.Traditional_way/6.Baby_Assignment/insurance_pre.csv")
```

```python
dataset = pd.get_dummies(dataset,dtype=int,drop_first=1)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'sex_male', 'bmi', 'children', 'smoker_yes']]
```

```python
dependent = dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

criteria = ['squared_error', 'absolute_error']
max_features_list = ['auto', 'sqrt', 'log2']
n_estimators_list = [10, 100]

results = []
sl_no = 1

for criterion in criteria:
    for max_features in max_features_list:
        for n in n_estimators_list:
            try:
                model = RandomForestRegressor(
                    criterion=criterion,
                    max_features=max_features,
                    n_estimators=n
                    #random_state=42
                )

                model.fit(x_train, y_train)
                y_pred = model.predict(x_test)

                r_value = r2_score(y_test, y_pred)

                results.append({
                    "SL.NO": sl_no,
                    "CRITERION": criterion,
                    "MAX FEATURES": max_features,
                    "N_ESTIMATORS": n,
                    "R VALUE": round(r_value, 4)
                })

                sl_no += 1

            except:
                pass


df_rf = pd.DataFrame(results)


df_rf["MAX FEATURES"] = df_rf["MAX FEATURES"].str.capitalize()

print(df_rf.to_string(index=False))
```

```python
best_row = df_rf.loc[df_rf["R VALUE"].idxmax()]
print("\nBest Combination:\n")
print(best_row)
```
