```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)

```

```python
x_train
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor
```

```python
from sklearn.metrics import r2_score
criteria = ['squared_error', 'absolute_error', 'friedman_mse']
max_features_list = ['auto', 'sqrt', 'log2']
splitters = ['best', 'random']
results = []
sl_no = 1

for criterion in criteria:
    for max_features in max_features_list:
        for splitter in splitters:
            try:
                model = DecisionTreeRegressor(
                    criterion=criterion,
                    max_features=max_features,
                    splitter=splitter
                   # random_state=42
                )

                model.fit(x_train, y_train)
                y_pred = model.predict(x_test)

                r_value = r2_score(y_test, y_pred)

                results.append({
                    "SL.NO": sl_no,
                    "CRITERION": criterion,
                    "MAX FEATURES": max_features,
                    "SPLITTER": splitter,
                    "R VALUE": round(r_value, 4)
                })

                sl_no += 1

            except:
                pass
df_dt = pd.DataFrame(results)

print(df_dt.to_string(index=False))
```

```python
best_row = df_dt.loc[df_dt["R VALUE"].idxmax()]
best_row
```
