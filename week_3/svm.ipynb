```python
dataset = pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/HOPE AI/WEEK-3-ML/1.Traditional_way/3.SVM/50_Startups.csv") #reading file from mounted gdrive
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
dependent = dataset[['Profit']]
```

```python
independent = dataset[['R&D Spend', 'Administration', 'Marketing Spend',
       'State_Florida', 'State_New York']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size = 1/3, random_state = 0)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel = "rbf",C=1000)
regressor.fit(X_train, y_train)
```

```python
regressor.intercept_
```

```python
regressor.n_support_
```

```python
regressor.support_
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```
