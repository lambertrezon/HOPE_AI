```python
class OddEven:
    def OddEven(num):
        if num % 2 == 0:
            print(f"The number {num} is Even.")
        else:
            print(f"The number {num} is Odd.")
```

```python
class EligibilityForMarriage:
    def Eligible(gender, age):
        if gender.lower() == "male":
            if age >= 21:
                print(f"Male, age {age}: Eligible for marriage.")
            else:
                print(f"Male, age {age}: Not eligible for marriage.")
        elif gender.lower() == "female":
            if age >= 18:
                print(f"Female, age {age}: Eligible for marriage.")
            else:
                print(f"Female, age {age}: Not eligible for marriage.")
        else:
            print("Invalid gender. Please specify 'male' or 'female'.")
```

```python
class FindPercent:
    def percentage(self, marks):
        for mark in marks:
            print(f"Subject {marks.index(mark) + 1}: {mark}")
        total_marks = sum(marks)
        percentage = (total_marks / (len(marks) * 100)) * 100
        return total_marks, percentage

marks = [98, 87, 95, 95, 93]

find_percent = FindPercent()
total, percent = find_percent.percentage(marks)

```

```python
class Triangle:
    def __init__(self, height, breadth, height1, height2):
        self.height = height
        self.breadth = breadth
        self.height1 = height1
        self.height2 = height2

    def area(self):
        return (self.height * self.breadth) / 2

    def perimeter(self):
        return self.height1 + self.height2 + self.breadth

# Create an instance of the Triangle class
triangle = Triangle(32, 34, 2, 4)

```
