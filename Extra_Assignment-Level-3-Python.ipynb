```python
age = int(input("Please enter your age: "))

if age < 13:
  print("Child")
elif age < 18:
  print("Teenager")
elif age < 60:
  print("Adult")
else:
  print("Senior Citizen")

```

```python
number = int(input("Enter a number: "))

if number > 0:
  print("The number is positive")
elif number < 0:
  print("The number is negative")
else:
  print("The number is zero")
```
