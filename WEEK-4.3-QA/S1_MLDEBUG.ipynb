```python
#q1
import pandas as pd

data = {'Feature1': ['10', '20', 'Thirty'],
        'Feature2': [5.5, 6.7, 8.9]}
df = pd.DataFrame(data)
print(data)
df['Feature1'] = pd.to_numeric(df['Feature1'], errors='coerce')
# ValueError: invalid literal for int() with base 10: 'Thirty'
# on executing line 7 while converting to int datatype the string "thirty" cannot be converted to int
# convert invalid data to valid one. Source : https://pandas.pydata.org/docs/reference/api/pandas.to_numeric.html
print("=============")
print(df)
```

```python
#q2
import pandas as pd


data = {'A': [1, 2, None],
        'B': [4, None, 6]}
#print(data)
df = pd.DataFrame(data)
#print(df.mean())
out_mean_value = df.mean()
df.fillna(out_mean_value)
# NAN cannot be considered as data. so use inplace flat in df.fillna method. source: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.fillna.html
print(df)
```

```python
#q8
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder

X_train = [[1, 2], [3, 4], [5, 6]]
y_train = ["yes", "no", "yes"]

encoder_final = LabelEncoder()
y_train_encoded_exc = encoder_final.fit_transform(y_train)

model = LogisticRegression()
#print("model -> ")
#print(model)
model.fit(X_train, y_train_encoded_exc)

print(model.predict([[2, 3]]))
#labels are in string, numeric is required when it comed to regression logistics
```

```python
#q9
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer

import numpy as np
X_train = np.array([[1, 2], [3, np.nan], [5, 6]])
y_train = np.array([10, 20, 30])

# data has missing values, nan cannot be considered for regression. values may be replaced with minimum mean value
imputer = SimpleImputer(strategy="mean")
X_train_imputed = imputer.fit_transform(X_train)
model = LinearRegression()
model.fit(X_train_imputed, y_train)


prediction = model.predict([[3, 4]])
print("Predicted Output:", prediction)
```

```python
#q11
from sklearn.tree import DecisionTreeClassifier

X_train = [[1, 2], [3, 4], [5, 6]]
y_train = ["spam", "ham", "spam"]  # model predicting wrongly as string cannot be considered for classifier. string to be converted to numeric

encoder = LabelEncoder()
y_train_fina = encoder.fit_transform(y_train)

model = DecisionTreeClassifier()
model.fit(X_train, y_train_fina)
print(model.predict([[2, 3]]))
```
